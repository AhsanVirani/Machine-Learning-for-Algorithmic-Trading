```python
import yfinance
import matplotlib.pyplot as plt
```

```python
btc = yfinance.download(tickers='BTC-USD')
```

```python
btc['max'] = btc['High'].rolling(20).max()
btc['min'] = btc['Low'].rolling(20).min()
btc['mid'] = (btc['max'] + btc['min']) / 2

# average of mid band
btc['mid_30'] = btc.mid.rolling(30).mean()
btc['mid_50'] = btc.mid.rolling(50).mean()
btc['mid_90'] = btc.mid.rolling(90).mean()

# # standard deviation from mid point
sd = btc['mid'].rolling(20).std(2)
btc['upper_band'] = btc['mid'] + sd
btc['lower_band'] = btc['mid'] - sd

# btc['upper_band'] = 
```

```python
btc[['Close', 'min', 'max', 'mid']][:500].dropna().plot(figsize=(14, 6))
btc[['Close', 'mid_30', 'mid_50', 'mid_90']][:500].dropna().plot(figsize=(14, 6))
btc[['Close', 'upper_band', 'lower_band', 'mid']][:500].dropna().plot(figsize=(14, 6))
```
